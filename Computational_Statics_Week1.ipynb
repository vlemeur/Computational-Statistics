## Problème 1.2

### a) Recherche du minimum de deux Lois normales

Soit $X \thicksim \mathcal{N}(\theta,\sigma^2)$ et Y $\thicksim \mathcal{N}(\mu,\tau^2)$ et  Z= X∧Y

Déterminons sa fonction de réparition en notant $\phi$ la fonction de répartition d'une loi normale :

$$ \begin{aligned} F(z)&=P(Z \le z)= P(X \le z) + P(Y \le z)- P(X \le z \cap Y \le z) \\
& = P(X \le z) + P(Y \le z) -P(X \le z)P(Y \le z) \\
& = \phi(\dfrac{z-\theta}{\sigma}) + \phi(\dfrac{z-\mu}{\tau}) - \phi(\dfrac{z-\theta}{\sigma})\phi(\dfrac{z-\mu}{\tau})
\end{aligned}$$

En dérivant on obtient alors la densité de probabilité voulue :

$$ \begin{aligned} f(z) & = \phi(\dfrac{z-\theta}{\sigma})\sigma^{-1} + \phi(\dfrac{z-\mu}{\tau}) - [\phi(\dfrac{z-\theta}{\sigma})\phi(\dfrac{z-\mu}{\tau})\sigma^{-1} + \phi(\dfrac{z-\theta}{\sigma})\phi(\dfrac{z-\mu}{\tau})\tau^{-1}] \\
& = [1-\phi(\dfrac{z-\mu}{\tau})]\phi(\dfrac{z-\theta}{\sigma})\sigma^{-1} + [1-\phi(\dfrac{z-\theta}{\sigma})]\phi(\dfrac{z-\mu}{\tau})\tau^{-1}
\end{aligned}$$

### b) Recherche du minimum d'une loi de Weibull et d'une constante $\omega$

Soit $X \thicksim \mathcal{We}(\alpha,\beta)$ et $\omega$ une constante et  Z= X∧$\omega$

On commence par déterminer la fonction de réparition de Z :

$$ \begin{aligned} F(z)&=P(Z \le z)= 1 - P(min(X,\omega) = 1- (1-P(X \le z))(1-P(\omega \le z)) \\
&= P(X \le z) + P(\omega \le z) -P(X \le z)P(\omega \le z)
\end{aligned}$$

On calcule directement la densité par dérivation :

$$ \begin{aligned} f(z)&= P'(X \le z) + P'(\omega \le z) -[P'(X \le z)P(\omega \le z)+P(X \le z)P'(\omega \le z)] \\
&= \alpha\beta z^{\alpha-1}exp(-\beta z^{\alpha}) + \delta_{\omega}(z) -[(\int_{-\infty}^{\omega}{\alpha\beta x^{\alpha-1}}exp(-\beta x^{\alpha})dx)\delta_{\omega}(z) + \alpha\beta z^{\alpha-1}exp(-\beta z^{\alpha})I_{\omega \le  z}] \\
\end{aligned}$$

Or $I_{\mathcal{R}} =1$ et $\int_{-\infty}^{+\infty}{\alpha\beta x^{\alpha-1}}exp(-\beta x^{\alpha})dx = 1 $ donc on peut réecrire les deux premiers termes de la quantité précédente : 

$$ \begin{aligned} f(z)&= \alpha\beta z^{\alpha-1}exp(-\beta z^{\alpha})I_{\mathcal{R}} + \delta_{\omega}(z)\int_{-\infty}^{+\infty}{\alpha\beta x^{\alpha-1}}exp(-\beta x^{\alpha})dx -[(\int_{-\infty}^{\omega}{\alpha\beta x^{\alpha-1}}exp(-\beta x^{\alpha})dx)\delta_{\omega}(z) + \alpha\beta z^{\alpha-1}exp(-\beta z^{\alpha})I_{\omega \le  z}] \\
\end{aligned}$$

On obtient enfin la formule souhaitée :

$$ \begin{aligned} f(z)&= \alpha\beta z^{\alpha-1}exp(-\beta z^{\alpha})I_{z \le \omega} + \int_{\omega}^{\infty}{\alpha\beta x^{\alpha-1}}exp(-\beta x^{\alpha})dx)\delta_{\omega}(z)
\end{aligned}$$

## Problème 1.7

On considère ici un n-échantillon $(Y_1,...,Y_n)$ i.i.d suivant une loi normale de densité : 

$$ \\ f(y) = \dfrac{1}{\sigma \sqrt{2\pi}} e^{\dfrac{-(y-\mu)^2}{2\sigma^2}}$$

On pose $ \theta =(\mu,\sigma^2) $ et on cherche à déterminer les équations du maximum de vraisemblance.

Calculons la vraisemblance de l'échantillon : $$ \\ L(\theta,y_1,...,y_n) = \prod_{i=1}^{n}{f(\theta,y_i)} $$

$$ \\ L(\theta,y_1,...,y_n) = \prod_{i=1}^{n}{\dfrac{1}{\sigma \sqrt{2\pi}} e^{\dfrac{-(y_i-\mu)^2}{2\sigma^2}}} 
= (2\pi\sigma^2)^{-\dfrac{n}{2}} e^{-\sum_{i=1}^{n}{\dfrac{(y_i^2+\mu^2-2x_i\mu)}{2\sigma^2}}}  $$

Déduisons à présent la log-vraisemblance de l'échantillon :  $$ \\ l(\theta,y_1,...,y_n)= \ln{L(\theta,y_1,...,y_n)} $$

$$  l(\theta,y_1,...,y_n)= -\dfrac{n}{2}\ln{(2\pi\sigma^2)} -\sum_{i=1}^{n}{\dfrac{y_i^2}{2\sigma^2}} -\dfrac{n\mu^2}{2\sigma^2} + \dfrac{\mu}{\sigma^2}\sum_{i=1}^{n}{y_i} $$

On dérive ensuite la log-vraisemblance par rapport à chaque variable de $ \theta $ :

$$\dfrac{\partial l(\theta,y_1,...,y_n)}{\partial \mu} = -\dfrac{n\mu}{\sigma^2} + \dfrac{1}{\sigma^2}\sum_{i=1}^{n}{y_i} \\
\dfrac{\partial l(\theta,y_1,...,y_n)}{\partial \sigma^2} = -\dfrac{n}{2\sigma^2}  +\dfrac{\sum_{i=1}^{n}{y_i^2}}{2\sigma^4} + \dfrac{n\mu^2}{2\sigma^4} -\dfrac{\mu}{\sigma^4}\sum_{i=1}^{n}{y_i}
$$

On égalise les deux quantités précédentes à 0 pour trouver l'estimateur de  maximum de vraisemblance pour $\theta$ :

$$ \dfrac{\partial l(\theta,y_1,...,y_n)}{\partial \mu} = 0 \Leftrightarrow \hat{\mu} = \dfrac{\sum_{i=1}^{n}{y_i}}{n}   = \bar{y}\\
\dfrac{\partial l(\theta,y_1,...,y_n)}{\partial \sigma^2} = 0 \Leftrightarrow  n\sigma^2 = \sum_{i=1}^{n}{y_i^2}+ n\mu^2 - 2\mu\sum_{i=1}^{n}{y_i} \Leftrightarrow \hat{\sigma^2}=\dfrac{\sum_{i=1}^{n}{(y_i-\mu)^2}}{n}
$$

En posant $ s^2=\dfrac{\sum_{i=1}^{n}{(y_i-\mu)^2}}{n}$ la variance empirique , on obtient bien l'estimateur de maximum de vraisemblance souhaité : $$ \hat{\theta} = (\bar{y},s^2) $$

## Problème 2.1

### a) Création de 1000 variables aléatoires et tracé d'un histogramme 

On génère 1000 variables aléatoires et uniforme et on trace l'histogramme de l'échantillon :

```r
par(mfrow = c(2,1))
X = runif(1000)
plot(X,main="Tracé du nuage de point issu de 1000 variables alétoires uniformes")
hist(X,freq=FALSE)
```

 On constate que l'histogramme obtenu conforte globalement la répartition sous forme de loi uniforme
 entre 0 et 1. Néanmoins, il est clair que cette réparition n'est pas tout à fait uniforme.

### b) Génération de n variables aléatoires et tracé des pairs pour autocorrélation

On trace pour chaque point de notre échantillon uniforme X[k+1] en fonction de X[k] :

```r
plot(X[1:999],X[2:1000],main="Tracé  X[k+1] et en fonction de X[k]")
```

Le tracé obtenu ne nous permet pas de conclure d'une corrélation entre les variables X[k] et X[k+1]. Le nuage de point est analogue à celui de la loi uniforme initiale. On constate néanmoins la présence de "grappes" ce qui remet à nouveau en question le caractère "parfaitement uniforme" de nos variables.

Pour vérifier l'indépendance, on décide de voir nos variables aléatoires uniformes comme une série temporelle pour laquelle nous allons tester l'indépendance entre deux termes consécutifs.

On effectue alors les tests "Box-Pierce" et "Ljung" pour tester l'indépendance entre X[k] et X[k+1] :

```r
Box.test(X, lag = 1)
Box.test (X, lag = 1, type = "Ljung")
```

La p-value élevée ici nous conforte dans l'hypothèse d'indépendance entre deux variables uniformes consécutives. 

## Problème 2.2

### a) Création de 1000 variables aléatoires binomiales à partir de lois uniformes

On commence par calculer les probabilités $ sk=P(X\le k)$ pour une loi binomiale en s'arrêtant dès que les probabilités dépassent $ p=0,99999 $

```r
p=0.2
n=25
P <-c()
k=0
pk=0
sk=0
while (sk<0.99999) {
    pk=choose(n,k)*p^k*(1-p)^(n-k)
    sk <- sk + pk
    P[k+1]=sk
    k <-k+1
}
P
```

On effectue l'approximation suivante : $ 0,99999 \simeq 1 $

On stocke dans un vecteur P nos différentes probabilités sk par ordre croissant.
On obtient  16 probabilités. Ainsi notre variable aléatoire binomiale pourra prendre les valeurs entières entre 0 et 15

On tire ensuite 1000 lois uniformes dans [0,1] :

```r
U = runif(1000)
```

On cherche à former 1000 variables aléatoires binomiales notées Bin à partir des ces variables aléatoires uniformes (notées U) grâce à la relation suivante : 
 
 $ s_k \le U \le s_{k+1} $ alors $Bin=k+1$

```r
aleagen <- function(u,P) {
    # Permet d'associer à une probabilité u le rang k dans une liste P croissante tel que u<P[k]
    # La fonction retourne le k associé 
    k=1
    repeat {
    if (u<P[k]) break
    k<-k+1
    }
    return(k-1)
}
    
```

```r
Bin <-c()
for (k in seq(1,length(U))) {
    a=U[k]
    Bin[k]=aleagen(a,P)
    }
    
```

Traçons l'histogramme de nos 1000 variables aléatoires binomiales obtenues : 

```r
hist(Bin,main="Tracé de l'histogramme de 1000 variables aléatoires binomiales",freq=FALSE)
```

Nous allons à présent tracer la densité de loi binomiale pour ces 15 valeurs et comparer la forme obtenue avec l'histogramme précédent

```r
plot(1:15,dbinom(1:15,n,p),main="Tracé de la densité d'une binomiale pour chaque valeur possible",type="p")
     
     
```

On constate que la forme de l'histogramme empirique et de la fonction de densité sont très proches ce qui conforte nos résultats.

Nous allons à présent comparer nos résultats avec la création de 1000 variables aléatoires binomiales avec la fonction dédiée de R :

```r
hist(rbinom(1000,n,p),main="Histogramme de 1000 v.a binomiales avec le générateur de R",freq=FALSE)
```

On constate une forme analogue à celle obtenue par notre construction empirique mais avec des différences de répartition notament pour les valeurs moyenne (X=4, X=5, X=6). Cet histogramme semble légèrement plus éloigné de la forme de la densité que notre histogramme empirique.

### b) Création de 5000 variables aléatoires logarithmiques à partir de lois uniformes et sa "densité"

Ici on prends p=0.2 la probabilité de notre variable aléatoire discrète a pour forme :
$ P(X=x) = \dfrac{-(1-p)^x}{xlog(p)} $

On applique la même méthode que pour la création de variable aléatoire binomiale

On  calcule alors les différentes probabilités $ lk=P(X\le k)$ qu'on stocke dans un vecteur L :

```r
L <-c()
p=0.2
k=1
pk=0
lk=0
while (lk<0.99999) {
    pk=(-(1-p)^k)/(k*log(p))
    lk <- lk + pk
    L[k]=lk
    k <-k+1
}
```

On effectue ici encore l'approximation  : $ 0,99999 \simeq 1 $

On obtient ici  39 probabilités donc notre variable aléatoire discrète  pourra prendre pour valeur les entiers positifs entre 1 et 39

On crée cette fois 5000 variables aléatoires uniformes :

```r
U=runif((5000))
```

Et on en déduit nos variables logarithmiques :

```r
Xlog <-c()
for (k in seq(1,length(U))) {
    a=U[k]
    Xlog[k]=aleagen(a,L)
    }
```

```r
hist(Xlog,main="Histogramme de 5000 v.a logarithmiques",freq=FALSE)
```

```r
plot(1:39,(-(1-p)^(1:39))/((1:39)*log(p)),type="p",main="Tracé de la densité de la loi pour chaque valeur possible")
```

Ici encore on obtient un tracé de densité théorique de forme analogue à notre histogramme empirique ce qui conforte notre construction.
